# **Data Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **1. Financial Review**

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/project/query-result (1).csv')
df
```

```python
fig = sns.catplot(data = df, x = 'order_month', y = 'pct_monthly_sales', kind = 'point')
fig.fig.set_size_inches(15, 6)

plt.grid(True, axis = 'y', alpha = 0.5)
plt.xticks(rotation = 45, fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('order_month', fontdict = {'family':'sans-serif', 'weight':'semibold', 'size':'large'}, labelpad = 20)
plt.ylabel('pct_monthly_sales', fontdict = {'family':'sans-serif', 'weight':'semibold', 'size':'large'}, labelpad = 20)

plt.axvline(6, color = '#C51100', linestyle = '-', linewidth = 3)
plt.axvline(9, color = '#C51100', linestyle = '--', linewidth = 3)
plt.axvline(11, color = '#C51100', linestyle = '-', linewidth = 3)
plt.axvspan(6, 11, facecolor = '#FAED09', alpha = 0.2)
plt.show()
```

## **2. Retention Review**

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/project/query-result (8).csv').set_index('first_order_month').drop(columns = 'cnt_month0')
df
```

```python
sns.set_theme(rc = {'figure.figsize':(20, 15)})
sns.heatmap(data = df, linewidths = 0.01, square = True, cmap = 'Blues', annot = True, annot_kws = {'size':13.5}, fmt = '.2f', vmax = 53.13)

plt.yticks(rotation = 45, fontsize = 13)
plt.xticks(rotation = 45, fontsize = 13)
plt.xlabel('pct_after_first_order_month', fontdict = {'family':'sans-serif', 'weight':'semibold', 'size':'large'}, labelpad = 20)
plt.ylabel('cohort_first_order_month', fontdict = {'family':'sans-serif', 'weight':'semibold', 'size':'large'}, labelpad = 20)

plt.axhline(7, color = '#C51100', linestyle = '-', linewidth = 4)
plt.axhline(8, color = '#C51100', linestyle = '-', linewidth = 4)
plt.title('Classic Retention', fontdict = {'fontsize':23, 'fontweight':'bold'}, pad = 30)
plt.show()
```
